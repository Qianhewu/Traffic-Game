```python
import numpy as np
import queue
N = 10 # number of players
f = 20 # frequency
AVG = 0.6
VAR = 0.1
theta = np.random.normal(AVG,VAR,N)
theta = np.minimum(np.maximum(theta,0),1)

def cost(x):
    return x


M = cost(1)



def runs(t,N,f):
    finished_t = np.zeros(N)
    queuing_n = 1
    finished = 0
    sorted_t = np.sort(t)
    index = np.argsort(t)
    
    q = queue.Queue()
    current = sorted_t[0]
    q.put(index[0])
    
    x = [0,current, current]
    y = [0,0,1]
    while finished < N:
        if q.qsize() > 0:
            current += 1/f
            finished += 1
            finished_t[q.get()] = current
            for i in range(queuing_n,N):
                if sorted_t[i] <= current:
                    q.put(index[i])
                    queuing_n += 1
            x.append(current)
            y.append(q.qsize())
        else:
            x.append(current)
            y.append(q.qsize())
            if queuing_n >= N:
                break
            current = sorted_t[queuing_n]
            q.put(index[queuing_n])
            queuing_n += 1
            x.append(current)
            y.append(q.qsize())
    return x,y,finished_t
    
# # print(theta)
# t= theta - 0.2
# x,y,finished_t = runs(t)
# # print(t)
# # print(finished_t)
# # print(theta > finished_t)
```

```python
class traffic_game():
    def __init__(N, f, M, AVG, VAR):
        self.N = N
        self.f = f
        self.M = M
        self.AVG = AVG
        self.VAR = VAR
        self.theta =  np.minimum(np.maximum(np.random.normal(AVG,VAR,N),0),1)
        self.t = np.random.uniform(0,theta, N)

    def runs(self):
        finished_t = np.zeros(self.N)
        queuing_n = 1
        finished = 0
        sorted_t = np.sort(self.t)
        index = np.argsort(self.t)
        q = queue.Queue()
        current = sorted_t[0]
        q.put(index[0])
        x = [0,current, current]
        y = [0,0,1]
        while finished < self.N:
            if q.qsize() > 0:
                current += 1/self.f
                finished += 1
                finished_t[q.get()] = current
                for i in range(queuing_n,self.N):
                    if sorted_t[i] <= current:
                        q.put(index[i])
                        queuing_n += 1
                x.append(current)
                y.append(q.qsize())
            else:
                x.append(current)
                y.append(q.qsize())
                if queuing_n >= self.N:
                    break
                current = sorted_t[queuing_n]
                q.put(index[queuing_n])
                queuing_n += 1
                x.append(current)
                y.append(q.qsize())
        return x,y,finished_t
    
    def compute_optimal(self)
        epsilon = 0.001
        opt = np.zeros(len(self.t))
        for i in range(len(opt)):
            c = self.t.copy()
            a = 0
            b = 1
            while b-a > epsilon:
                c[i] = (a+b)/2
                x,y,finish = runs(c,self.N,self.f)
                if finish[i]>theta[i]:
                    b = (a+b)/2
                else:
                    a = (a+b)/2
            opt[i] = (a+b)/2
        return opt
    
def visualize(x,y):
#     plt.plot(x, y, 'ro-', color='#4169E1', alpha=0.8, linewidth=1)
    plt.plot(x, y, 'ro-', color='#4169E1', alpha=0.8, linewidth=1)
    plt.show()
```
